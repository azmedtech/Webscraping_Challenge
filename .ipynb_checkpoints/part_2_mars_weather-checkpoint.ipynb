### Step 3: Store the Data

Assemble the scraped data into a Pandas DataFrame. The columns should have the same headings as the table on the website. Hereâ€™s an explanation of the column headings:

* `id`: the identification number of a single transmission from the Curiosity rover
* `terrestrial_date`: the date on Earth
* `sol`: the number of elapsed sols (Martian days) since Curiosity landed on Mars
* `ls`: the solar longitude
* `month`: the Martian month
* `min_temp`: the minimum temperature, in Celsius, of a single Martian day (sol)
* `pressure`: The atmospheric pressure at Curiosity's location

```python
# Create a Beautiful Soup Object
html = browser.html
soup = BeautifulSoup(html, 'html.parser')

# Extract all rows of data, excluding header row
table = soup.find_all('tr', class_='data-row')

# Create an empty list to store the table data
mars_table_data = []

# Loop through the rows to extract data
for row in table:
    # Create a new list for each row of data
    ind_row_list = []
    
    for item in row:
        clean_item = item.get_text()
        if (clean_item) != '\n':
            ind_row_list.append(clean_item)
               
    # Add the row data dictionary to the list of table data
    mars_table_data.append(ind_row_list)

# Print the list of dictionaries containing the table data
mars_table_data
```

```python
# Identify the column headers
column_names =["id","terrestrial_date","sol","ls","month","min_temp","pressure"]
```

```python
# Create a Pandas DataFrame by using the list of rows and a list of the column names
mars_data_df = pd.DataFrame(mars_table_data, columns=column_names)

# Set the 'id' column as the index
mars_data_df.set_index('id', inplace=True)
mars_data_df
```

### Step 4: Prepare Data for Analysis

Examine the data types that are currently associated with each column. If necessary, cast (or convert) the data to the appropriate `datetime`, `int`, or `float` data types.

  > **Hint** You can use the Pandas `astype` and `to_datetime` methods to accomplish this task.


```python
# Examine data type of each column
mars_data_df.info()
```

```python
# Change data types for data analysis
mars_data_df['terrestrial_date'] = pd.to_datetime(mars_data_df['terrestrial_date'])
mars_data_df['sol'] = mars_data_df['sol'].astype(int)
mars_data_df['ls'] = mars_data_df['ls'].astype(int)
mars_data_df['month'] = mars_data_df['month'].astype(int)
mars_data_df['min_temp'] = mars_data_df['min_temp'].astype(np.float64)
mars_data_df['pressure'] = mars_data_df['pressure'].astype(np.float64)
```

```python
# Confirm type changes were successful by examining data types again
mars_data_df.info()
```

```python
# 1. How many months are there on Mars?
unique_mars_months_count = mars_data_df['month'].nunique()
print(f"{unique_mars_months_count} unique months exist on Mars.")
```

```python
# 2. How many Martian days' worth of data are there?
martian_days_count = mars_data_df['sol'].nunique()

print(f"There are {martian_days_count} Martian days' worth of data in the scraped dataset.")
```

```python
# 3. What is the average low temperature by month?
average_low_temp_by_month = mars_data_df.groupby('month')['min_temp'].mean()

print("Average Low Temperature by Month:")
print(average_low_temp_by_month)
```

```python
# Plot the average temperature by month

# Define the x-axis values (months)
months = range(1, 13)

# Create the histogram plot
plt.figure(figsize=(10, 6))
plt.bar(months, average_low_temp_by_month, color='skyblue')

# Set the labels and title
plt.xlabel('Month')
plt.ylabel('Temperature in Celsius')
plt.title('Average Temperature by Martian Month')

# Set the y-axis range from 0 to -80
plt.ylim(-100, 0)

# Set the x-axis range from 1 to 12
plt.xlim(0, 13)

# Show the plot
plt.show()
```

```python
# Identify the coldest and hottest months in Curiosity's location

# Convert series to dataframe
avg_lt_by_month = pd.DataFrame(average_low_temp_by_month)
avg_lt_by_month.head()

# Find the coldest month
coldest_month = avg_lt_by_month.loc[avg_lt_by_month['min_temp'].idxmin()]

# Find the hottest month
hottest_month = avg_lt_by_month.loc[avg_lt_by_month['min_temp'].idxmax()]

print(f"The coldest month at Curiosity's location is Month {coldest_month}")
print(f"The hottest month at Curiosity's location is Month {hottest_month}")
```

```python
# Define the bucket size in Earth days
bucket_size = 250

# Sort the DataFrame by 'terrestrial_date' in ascending order
mars_data_df_sorted = mars_data_df.sort_values('terrestrial_date')

# Initialize variables to store the highest 'min_temp' for each bucket and the corresponding bucket start dates
bucket_max_temps = []
bucket_start_dates = []

# Iterate over the dataset in buckets of 250 Earth days
for i in range(0, len(mars_data_df_sorted), bucket_size):
    bucket_data = mars_data_df_sorted.iloc[i:i+bucket_size]
    if not bucket_data.empty:
        max_temp = bucket_data['min_temp'].max()
        start_date = bucket_data['terrestrial_date'].iloc[0]
        bucket_max_temps.append(max_temp)
        bucket_start_dates.append(start_date)

# Create a DataFrame with the bucket start dates and highest 'min_temp' values
buckets_df = pd.DataFrame({'start_date': bucket_start_dates, 'max_temp': bucket_max_temps})

# Find the three buckets with the highest 'min_temp' values
highest_three_buckets = buckets_df.nlargest(3, 'max_temp')

# Order the dates in ascending order
highest_three_buckets = highest_three_buckets.sort_values('start_date')

# List the 'terrestrial_date' for each of the three highest 'min_temp' values
print("Terrestrial dates for the three highest 'min_temp' values:")
for index, row in highest_three_buckets.iterrows():
    print(row['start_date'])

# Calculate the number of days between the first and second date (absolute value)
days_between_first_second = abs((highest_three_buckets.iloc[1]['start_date'] - highest_three_buckets.iloc[0]['start_date']).days)
print(f"Number of days between the first and second date: {days_between_first_second}")

# Calculate the number of days between the second and third date
days_between_second_third = abs((highest_three_buckets.iloc[2]['start_date'] - highest_three_buckets.iloc[1]['start_date']).days)
print(f"Number of days between the second and third date: {days_between_second_third}")

```
